```python
%pip install --quiet openai pandas python-dotenv

import os
from dotenv import load_dotenv
from openai import OpenAI
import pandas as pd

load_dotenv()  # Lê o .env na raiz do projeto
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

```

```python
def classificar_sentimento(texto):
    prompt = f"""Classifique o sentimento do feedback abaixo como Positivo, Neutro ou Negativo.
    Responda apenas com uma dessas palavras.
    
    Feedback: {texto}"""

    resposta = client.responses.create(
        model="gpt-4.1",  # ou o modelo que você tiver acesso
        input=prompt,
        
    )

    return resposta.output_text.strip()
```

```python
df = pd.read_csv('data/reviews.csv')

df["sentimento"] = [
    classificar_sentimento(review)
    for review in df["reviewText"]
]

df.to_csv("data/reviews.csv", index=False)
```
